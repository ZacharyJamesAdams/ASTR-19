Plotting a 2 datasets on single `Figure` as two separate `Axes` using `matplotlib`.  

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.0,1.0,100)
```

Defining functions two get both independent variables.

```python
def sin(x):
    return np.sin(x)
```

```python
def cos(x):
    return np.cos(x)
```

In this cell:  
1. Creating a `Figure` and add to it a 2 `Axes` in a 2x1 configuration
2. Plotting sin(x) vs. x in the left panel
3. Plotting cos(x) vs. x in the right panel
4. Changing panel titles to  `sin(x)` and `cos(x)` respectively
5. Saving the figure as a PDF.

```python
fig, ax = plt.subplots(1, 2, figsize=[i*4 for i in(2,1)])

ax[0].plot(x,sin(x))
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax0_title_text = ax[0].set_title('sin(x)')

ax[1].plot(x,cos(x))
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax1_title_text = ax[1].set_title('cos(x)')

fig.savefig('plot_prompt8.pdf', bbox_inches='tight', dpi=400)
```
