Create a Jupyter notebook, import matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Use numpy to pull 1000 random numbers distributed uniformly between [0,1].

```python
sample = np.random.uniform(0.0,1.0,1000)
```

1. Histogram the random numbers into 100 bins
2. plot the histogram
3. Label your axes
4. save the figure as a PDF.

```python
fig, ax = plt.subplots(1,1,figsize = (4,4))
ax.hist(sample,100)
ax.set_xlabel('x')
ax.set_ylabel('counts')
fig.savefig('plot_prompt9.pdf', bbox_inches = 'tight', dpi = 400)
```
