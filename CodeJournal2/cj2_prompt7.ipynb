Plotting a single-variable data set using `matplotlib`.  

```python
import numpy as np
import matplotlib.pyplot as plt
```

Defining the independent variable, x, for our data.

```python
x = np.linspace(0.0, 1.0, 100)
```

Defining a function to get the dependent variable given x.

```python
def get_exp(arr):
    return np.exp(arr)
```

In this cell:  
1. Assigning dependent variable to y
2. Setting up a `Figure` instance
3. Adding an `Axes` to that `Figure`
4. Setting the axis labels
5. Saving data in vectorized form (.pdf)

```python
y = get_exp(x)
fig,ax = plt.subplots(1, 1, figsize = (4, 4))
ax.plot(x, y)
ax.set_xlabel('Time [milliseconds]')
ax.set_ylabel('Awesomeness')
plt.savefig('plot_prompt7.pdf', bbox_inches='tight', dpi = 400)
```
