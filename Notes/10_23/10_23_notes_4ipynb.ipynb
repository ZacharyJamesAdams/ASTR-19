```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(119)

npoints = 50

x = np.linspace(0,10.,npoints)

m = 2.0
b = 1.0
sigma = 2.0

y = m*x + b + np.random.normal(scale=sigma, size=npoints)
y_err = np.full(npoints,sigma)
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.errorbar(x,y,sigma,fmt='o')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.errorbar(x,y,sigma,fmt='o', label = 'data')

m_fit, b_fit = np.poly1d(np.polyfit(x,y,1,w=1./y_err))
print(m_fit,b_fit)
y_fit = m_fit*x + b_fit

ax.plot(x,y_fit, label = 'fit')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(loc = 2,frameon=False)
```
