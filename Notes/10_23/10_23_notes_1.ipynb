```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import axes3d
```

```python
x = np.linspace(0, 1 * np.pi,200)
y = np.linspace(0, 2 * np.pi,200)
x,y = np.meshgrid(x,y)
z = -np.sin(x)*np.sin(y)
print(f'{z.shape = }')
# plt.imshow(z, cmap = cm.turbo)
ax = plt.imshow(z, cmap = cm.turbo, origin = 'lower') #lower flips origin to be bottom left vs top left as in image formats
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111,projection='3d')

ax.plot_wireframe(x, y, z, rstride=5, cstride=5, linewidth = .5,color = '0.25')
ax.view_init(20) # rotate view angle to 20 degrees
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111,projection='3d')

ax.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.coolwarm, alpha = 0.7)
ax.plot_wireframe(x, y, z, rstride=5, cstride=5, linewidth = .5,color = '0.5')
cset = ax.contourf(x,y,z, zdir='z', cmap = cm.coolwarm, offset= -1)
ax.view_init(10) # rotate view angle to 20 degrees
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.zaxis.line.set_lw(0.0)
ax.set_zticks([])
```
