```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches
```

```python
fig = plt.figure(figsize=(6,6))

ax = fig.add_subplot(1,1,1,xticks = [], yticks=[])
ax.set_xlim([0,1])
ax.set_ylim([0,1])

rectangle = patches.Rectangle((0.5, 0.5), 0.4, 0.4, edgecolor = 'None', facecolor = 'blue', alpha = 0.5)
ax.add_patch(rectangle)

circle = patches.Circle((0.6, 0.6), 0.3, edgecolor = 'None', facecolor = 'red', alpha = 0.5)
ax.add_patch(circle)

ellipse = patches.Ellipse((0.8, 0.8), 0.3, 0.4, edgecolor = 'None', facecolor = 'blue', alpha = 0.5)
ax.add_patch(ellipse)
ax.set_facecolor('white')

```
