```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
def func(t):
    r = 1.0
    x = r*np.cos(t)
    y = r*np.sin(t)
    return x,y
```

```python
t = np.arange(13)*np.pi/6
x,y = func(t)
print(t)
```

```python
fig,ax =plt.subplots(1,1, figsize = (5,5))
print(f'Figure dpi = {fig.get_dpi()}')
print(f'Figure res = {[fig.get_dpi()*i for i in (5.,5.)]}')
ax.axhline(0)
ax.axvline(0)
ax.set_aspect(1)
line = ax.plot(x,y)
scatter = ax.scatter([1,0,-1,0],[0,1,0,-1])
```

```python
fig,ax =plt.subplots(1,1, figsize = (5,5))
ax.set_aspect(1)
ax.spines[['top','right']].set_visible(False) 
# ax.spines[['left','bottom']].set_position(('data',0)) 
ax.spines[['left','bottom']].set_position(('axes',0.5)) 
line = ax.plot(x,y)
scatter = ax.scatter([1,0,-1,0],[0,1,0,-1])
```

```python
fig,ax =plt.subplots(1,1, figsize = (5,5))
ax.set_aspect(1)
ax.spines[['top','right']].set_position(('axes',0.5)) 
line = ax.plot(x,y)
scatter = ax.scatter([1,0,-1,0],[0,1,0,-1])
```

```python
x, y = func(np.linspace(0, 2*np.pi, 12, endpoint=False))
a, b = func(np.linspace(0, 2*np.pi, 13, endpoint=True))

print(x - a[:-1], x[0]-a[-1])

print(y - b[:-1], y[0]-b[-1])
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.set_aspect(1)
ax.spines[['top', 'right']].set_position(('axes', 0.5))
line = ax.plot(x, y)
scatter = ax.scatter([1, 0, -1,0],[0,1,0,-1])
```
